```python
import tkinter as tk
from tkinter import messagebox

class QuizApp(tk.Tk):
    def __init__(self):
        super().__init__()
        self.title("Shukriyyah Quiz Application")
        self.geometry("500x400")
        self.config(bg="lightgreen")

        self.subjects = {
            "Math": [
                {"question": "What is 5 + 7?", "options": ["10", "11", "12", "13"], "answer": "12"},
                {"question": "What is 8 * 3?", "options": ["21", "24", "27", "30"], "answer": "24"},
                {"question": "What is 199 * 19?", "options": ["1873", "8317", "3781", "3871"], "answer": "3781"},
                {"question": "If a car travels 250 miles in 5 hours, how many miles does it travel per hour?", "options": ["25", "50", "75", "100"], "answer": "50"},
                {"question": "What is the area of a circle with a radius of 4cm?", "options": ["16.00cm^2", "12.56cm^2", "20.00cm^2", "25.00cm^2"], "answer": "16.00 cm^2"},
                {"question": "What is 15 * 7?", "options": ["105", "151", "125", "153"], "answer": "105"},
                {"question": "What is 100 // 20 ?", "options": ["10", 5, "15", "7"], "answer": "5"},
                {"question": "What is 15 + 17?", "options": ["10", "11", "12", "32"], "answer": "32"},
                {"question": "What is 99 - 19?", "options": ["20", "80", "8", "13"], "answer": "80"},
                {"question": "What is 105 // 5?", "options": ["21", "11", "12", "13"], "answer": "21"},
],
                
            
            "English": [
                {"question": "What is the synonym of 'happy'?", "options": ["Sad", "Joyful", "Angry", "Tired"], "answer": "Joyful"},
                {"question": "What is the plural of 'Woman'?", "options": ["Women", "Woman", "Girl", "Lady"], "answer": "Women"},
                {"question": "What is the opposite of 'Boy'?", "options": ["Man", "Girl", "He", "Him"], "answer": "Girl"},
                {"question": "What is the singular of 'Ladies'?", "options": ["Lady", "Ladi", "Ledy", "Girl"], "answer": "Lady"},
                {"question": "What is the plural of 'Is'?", "options": ["ask", "As", "She", "Are"], "answer": "Are"},
                {"question": "What is the antonym of 'Cold'?", "options": ["Hot", "Warm", "Cool", "Freezing"], "answer": "Hot"},
                {"question": "What is the present tense of 'Eat'?", "options": ["Eats", "Ate", "Eating", "Eat"], "answer": "Eat"},
                {"question": "Which of these is the opposite of 'Clean'?", "options": ["Dirty", "Happy", "Fast", "Big"], "answer": "Dirty"},
                {"question": "What is the past tense of 'Write'?", "options": ["Writes", "Wrote", "Writing", "Written"], "answer": "Wrote"},
                {"question": "What is the meaning of the word 'Together'?", "options": ["Alone", "Seperated", "United", "Far"], "answer": "United"}
                    

            ],
                "Current Affairs": [
                {"question": "Who is the current President of the United States?", "options": ["Joe Biden", "Donald Trump", "Barack Obama", "George Bush"], "answer": "Joe Biden"},
                {"question": "Who is the current Governor of Lagos State?", "options": ["Babtunde Raji Fashola", "Oluremi Tinubu", "Babajide Sanwo-olu", "Bola Ahmed Tinubu"], "answer": "Babajide Sanwo-olu"},
                {"question": "Who is the current vice president of Nigeria?", "options": ["Oluremi Tinubu", "Gbajabiamila", "Kashim Shettima", "Adelabu Ibrahim"], "answer": "Kashim Shettima"},
                {"question": "Who formed the first political party in Nigeria?", "options": ["Nnamdi Azikwe", "Herbert Macaulay", "Olusegun Obasanjo", "MkO Abiola"], "answer": "Herbert Macaulay"},
                {"question": "Which country hosted the 2022 Winter Olympics?", "options": ["Germany", "USA", "Portugal", "China"], "answer": "China"},
                {"question": "What is the name of the largest planet in our solar system?", "options": ["Jupiter", "Mars", "Earth", "Century"], "answer": "Jupiter"},
                {"question": "What is the name of the major river that runs through Nigeria?", "options": ["River Niger", "River Benue", "River Ogun", "River Osun"], "answer": "River Niger"},
                {"question": "What is the name of the Nigerian Football Team?", "options": ["Super Eagles", "Super Falcons", "Golden Eaglets", "Flying Eagles"], "answer": "Super Eagles"},
                {"question": "What is the name of the Nigerian currency?", "options": ["Dollar", "Pound", "Naira", "kobo"], "answer": "Naira"},
                {"question": "Who is the current President of Nigeria?", "options": ["Muhammadu Buhari", "Goodluck Jonathan", "Olusegun Obasanjo", "Bola Ahmed Tinubu"], "answer": "Bola Ahmed Tinubu"},
                
                
            ],
                "Football": [
                {"question": "Which team has won the most Champions League titles?", "options": ["Real Madrid", "AC Milan", "Liverpool", "Bayern Munich"], "answer": "Real Madrid"},
                {"question": "Who scored the fastest hat-trick in Champions League history?", "options": ["Cristiano Ronaldo", "Lionel Messi", "Robert Lewandowski", "Sérgio Agüero"], "answer": "Robert Lewandowski"},
                {"question": "Which player has won the most Ballon d'Or awards?", "options": ["Cristiano Ronaldo", "Lionel Messi", "Diego Maradona", "Johan Cruyff"], "answer": "Lionel Messi"},
                {"question": "Which team won the 2018 FIFA World Cup?", "options": [" France", "Germany", "Brazil", "Argentina"], "answer": "France"},
                {"question": "Which team has won the most FA Cups in England?", "options": ["Machested United", "Liverpool", "Arsenal", "Chelsea"], "answer": "Arsenal"},
                {"question": "Which player has scored the most hat-tricks in the Premier League?", "options": ["Sergio Agüero", "Robbie Fowler", "Thierry Henry", "Harry Kane"], "answer": "Sergio Agüero"},
                {"question": "Which team won the Champions League in 2020?", "options": ["Arsenal", "Bayern Munich", "Liverpool", "Chealsea"], "answer": "Bayern Munich"},
                {"question": "Who is Arsenal's all-time leading scorer in the Premier League?", "options": ["Ian Wright", "Robin van Persie", "Olivier Giroud", "Thierry Henry"], "answer": "Thierry Henry (175 goals)"},
                {"question": "Which team has appeared in the most Champions League finals?", "options": ["Real Madrid", "AC Milan", "Liverpool", "Bayern Munich"], "answer": "Real Madrid"},
                {"question": "Which team won the first-ever Champions League title in 1993?", "options": ["Marseille", "AC Milan", " Barcelona", "Manchester United"], "answer": "Marseille"},
            ]
        }
                
        
        self.current_subject = None
        self.current_question = 0
        self.score = 0

        self.subject_label = tk.Label(self, text="Select a subject:", bg="light green", font=("Arial", 14))
        self.subject_label.pack(pady=20)

        self.subject_var = tk.StringVar(value="")
        self.subject_menu = tk.OptionMenu(self, self.subject_var, *self.subjects.keys())
        self.subject_menu.pack(pady=10)

        self.start_button = tk.Button(self, text="Start Quiz", command=self.start_quiz, bg="Purple")
        self.start_button.pack(pady=20)

        self.question_label = tk.Label(self, text="", bg="lightgreen", font=("Arial", 14))
        self.question_label.pack(pady=20)

        self.option_vars = [tk.StringVar(value="") for _ in range(4)]
        self.option_buttons = [tk.Radiobutton(self, text="", variable=self.option_vars[i], value="", bg="lightgreen", font=("Arial", 12)) for i in range(4)]
        for button in self.option_buttons:
            button.pack(anchor="w")
        self.submit_button = tk.Button(self, text="Submit", command=self.check_answer, bg="orange")
        self.submit_button.pack(pady=20)

        self.next_button = tk.Button(self, text="Next", command=self.next_question, bg="yellow")
        self.next_button.pack(pady=10)

        self.result_label = tk.Label(self, text="correct!", bg="green", font=("Arial", 12))
        self.result_label.pack(pady=10)

        self.hide_quiz_elements()

    def hide_quiz_elements(self):
        self.question_label.pack_forget()
        for button in self.option_buttons:
            button.pack_forget()
        self.submit_button.pack_forget()
        self.next_button.pack_forget()
        self.result_label.pack_forget()
    def show_quiz_elements(self):
        self.question_label.pack(pady=20)
        for button in self.option_buttons:
            button.pack(anchor="w")
        self.submit_button.pack(pady=20)
        self.next_button.pack(pady=10)
        self.result_label.pack(pady=10)

    def start_quiz(self):
        self.current_subject = self.subject_var.get()
        if self.current_subject:
            self.current_question = 0
            self.score = 0
            self.load_question()
            self.show_quiz_elements()
            self.subject_label.pack_forget()
            self.subject_menu.pack_forget()
            self.start_button.pack_forget()
        else:
            messagebox.showwarning("Warning", "Please select a subject.")
    def load_question(self):
        question_data = self.subjects[self.current_subject][self.current_question]
        self.question_label.config(text=question_data["question"])
        for i, option in enumerate(question_data["options"]):
            self.option_buttons[i].config(text=option, value=option)
            self.option_vars[i].set("")

    def check_answer(self):
        selected_option = next((var.get() for var in self.option_vars if var.get()), None)
        if selected_option:
            correct_answer = self.subjects[self.current_subject][self.current_question]["answer"]
            if selected_option == correct_answer:
                self.score += 1
                self.result_label.config(text="Correct!", fg="white")
            else:
                self.result_label.config(text=f"Incorrect! The correct answer is {correct_answer}.", fg="white")
        else:
            messagebox.showwarning("Warning", "Please select an option.")

    def next_question(self):
        self.current_question += 1
        if self.current_question < 10:
            self.load_question()
            self.result_label.config(text="")
        else:
            self.show_final_score()
            self.start_quiz()
            

    def show_final_score(self):
        messagebox.showinfo("Quiz Completed", f"Your final score is {self.score}/10")
         #self.destroy()

if __name__ == "__main__":
    app = QuizApp()
    app.mainloop()

```
